**1. Yêu cầu bài toán:**
Trong file `payments.csv`, tìm kế hoạch trả góp (số kỳ) có giá trị thanh toán trung bình trên mỗi đơn hàng là cao nhất.

**2. Phương pháp tiếp cận & Suy luận:**
* **Bước 1:** Tải dữ liệu các giao dịch thanh toán từ file `payments.csv`.
* **Bước 2:** Sử dụng phương pháp gom nhóm (`groupby`) dữ liệu dựa trên cột số kỳ trả góp.
* **Bước 3:** Tính giá trị trung bình (`mean`) của cột giá trị thanh toán cho từng nhóm.
* **Bước 4:** Sắp xếp kết quả theo thứ tự giảm dần để tìm ra số kỳ trả góp đứng top 1.

**3. Lý do chọn phương pháp:**
* **Sử dụng `groupby` và `mean()`:** Đây là bài toán tổng hợp thống kê cơ bản. Ta cần biết "giá trị đại diện" của một đơn hàng khi khách chọn trả 1 cục so với trả nhiều kỳ. Hàm `mean()` là phù hợp nhất để san phẳng các biến động (outliers) và đưa ra con số trung bình phản ánh đúng hành vi chi tiêu của từng nhóm.

```python
# Import thư viện
import pandas as pd
from google.colab import drive

# Kết nối với Google Drive
drive.mount('/content/drive')
```

**Thực hiện Bước 1 & 2: Tải dữ liệu và Kiểm tra cấu trúc**

Ta nạp dữ liệu thanh toán vào và kiểm tra xem tên cột lưu số kỳ trả góp và giá trị thanh toán chính xác là gì.

```python
# 1. Khai báo đường dẫn
file_path = '/content/drive/MyDrive/datathon-2026-round-1/payments.csv'

# 2. Đọc dữ liệu
df_payments = pd.read_csv(file_path)

# Hiển thị 5 dòng đầu tiên
# (Thường tên cột sẽ là 'payment_installments' và 'payment_value')
df_payments.head()
```

**Thực hiện Bước 3 & 4: Tính toán và Xếp hạng**

Gom nhóm dữ liệu theo số kỳ trả góp, tính trung bình số tiền khách đã trả, sau đó tìm ra đỉnh cao nhất.

```python
# Tính giá trị trung bình theo số kỳ trả góp (cột 'installments')
avg_payment_by_installment = df_payments.groupby('installments')['payment_value'].mean().sort_values(ascending=False)

print("Giá trị thanh toán trung bình theo số kỳ trả góp:")
# Chỉ in ra 15 dòng đầu để tránh bảng quá dài
print(avg_payment_by_installment.head(15))

# Lấy số kỳ trả góp mang lại giá trị cao nhất
best_plan = avg_payment_by_installment.idxmax()
best_value = avg_payment_by_installment.max()

print(f"\n--> Kế hoạch trả góp có giá trị trung bình cao nhất là: {best_plan} kỳ (với {best_value:.2f})")
```
