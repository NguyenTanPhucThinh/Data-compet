**1. Yêu cầu bài toán:**
Xác định phương thức thanh toán (`payment_method`) được sử dụng nhiều nhất trong các đơn hàng có trạng thái là đã hủy (thường là `canceled` hoặc `cancelled` trong cột `order_status`).

**2. Phương pháp tiếp cận & Suy luận:**
* **Bước 1:** Tải dữ liệu từ file `orders.csv`.
* **Bước 2:** Lọc bảng dữ liệu để chỉ giữ lại những đơn hàng bị hủy.
* **Bước 3:** Đếm tần suất xuất hiện của từng phương thức thanh toán trên tập dữ liệu vừa lọc.
* **Bước 4:** Lấy phương thức có số lượng đếm được cao nhất.

**3. Lý do chọn phương pháp:**
* **Lọc trước khi đếm (Boolean Indexing):** Thay vì đếm toàn bộ rồi mới lọc, ta dùng điều kiện lọc ngay từ đầu để loại bỏ hàng trăm ngàn đơn hàng giao thành công hoặc hoàn trả. Việc này giúp giảm tải bộ nhớ và tăng tốc độ xử lý.
* **Sử dụng `value_counts()`:** Đây là hàm đếm tần suất danh mục (categorical data) được tối ưu hóa tốt nhất trong Pandas, tự động sắp xếp kết quả từ cao xuống thấp mà không cần viết thêm hàm sắp xếp.

```python
# Import thư viện
import pandas as pd
from google.colab import drive

# Kết nối với Google Drive
drive.mount('/content/drive')
```

**Thực hiện Bước 1 & 2: Tải và Lọc dữ liệu**

Để đảm bảo không bị lỗi chính tả (giữa tiếng Anh Anh và Anh Mỹ), ta sẽ kiểm tra xem trạng thái hủy trong dữ liệu được ghi chính xác là gì trước khi lọc.

```python
# 1. Khai báo đường dẫn
file_path = '/content/drive/MyDrive/datathon-2026-round-1/orders.csv'

# 2. Đọc dữ liệu
df_orders = pd.read_csv(file_path)

# Lấy các trạng thái đơn hàng hiện có để biết từ khóa chính xác
unique_statuses = df_orders['order_status'].unique()
print(f"Các trạng thái đơn hàng trong dữ liệu: {unique_statuses}")

# Lọc dữ liệu (Giả sử từ khóa là 'canceled', hàm str.contains giúp bao quát cả 'canceled' và 'cancelled')
df_cancelled = df_orders[df_orders['order_status'].str.contains('cancel', case=False, na=False)]

print(f"\nSố lượng đơn hàng bị hủy tìm thấy: {len(df_cancelled)}")
```

**Thực hiện Bước 3 & 4: Thống kê và Chốt kết quả**

Áp dụng lệnh đếm `value_counts()` lên cột phương thức thanh toán của tập dữ liệu vừa lọc để tìm ra "thủ phạm" hàng đầu.

```python
# Đếm tần suất phương thức thanh toán
payment_counts = df_cancelled['payment_method'].value_counts()

print("Thống kê phương thức thanh toán của các đơn hàng bị hủy:")
print(payment_counts)

# Lấy phương thức phổ biến nhất
top_method = payment_counts.idxmax()
top_count = payment_counts.max()

print(f"\n--> Phương thức thanh toán phổ biến nhất khi hủy đơn là: '{top_method}' (với {top_count} đơn)")
```
